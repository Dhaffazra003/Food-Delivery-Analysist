# -----------------------------------------
# DATA CLEANING
# -----------------------------------------


```python
print("=" * 50)
print("DATASET SHAPE")
print("=" * 50)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

print("\n")

print("=" * 50)
print("COLUMN NAMES")
print("=" * 50)

print(df.columns)

print("\n")

print("=" * 50)
print("DATA TYPES")
print("=" * 50)

print(df.dtypes)
```

# =========================================
# 2. HANDLE DUPLICATES
# =========================================

```python
print("\n")
print("=" * 50)
print("DUPLICATE CHECK")
print("=" * 50)

duplicate_count = df.duplicated().sum()

print(f"Duplicate Rows Before Cleaning : {duplicate_count}")

df = df.drop_duplicates()

duplicate_count_after = df.duplicated().sum()

print(f"Duplicate Rows After Cleaning  : {duplicate_count_after}")
```

# =========================================
# 3. HANDLE DATETIME COLUMN
# =========================================


```python
print("\n")
print("=" * 50)
print("DATETIME CONVERSION")
print("=" * 50)

df["Waktu_Transaksi"] = pd.to_datetime(
    df["Waktu_Transaksi"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)

datetime_missing = df["Waktu_Transaksi"].isnull().sum()

print(f"Invalid Datetime Rows : {datetime_missing}")
```

# =========================================
# 4. HANDLE MISSING VALUES
# =========================================

```python
print("\n")
print("=" * 50)
print("MISSING VALUES BEFORE CLEANING")
print("=" * 50)

print(df.isnull().sum())
```

# -----------------------------------------
# NUMERICAL COLUMNS
# -----------------------------------------

```python
numerical_columns = [
    "Jarak_Kirim_KM",
    "Rating_Pelanggan"
]

for column in numerical_columns:

    median_value = df[column].median()

    df[column] = df[column].fillna(median_value)
```

# -----------------------------------------
# TEXT COLUMN
# -----------------------------------------

```python
df["Ulasan_Teks"] = df["Ulasan_Teks"].fillna(
    "No Review"
)
```

# =========================================
# 5. FEATURE ENGINEERING
# =========================================

```python
print("\n")
print("=" * 50)
print("FEATURE ENGINEERING")
print("=" * 50)

# Extract hour
df["Jam_Transaksi"] = df["Waktu_Transaksi"].dt.hour

# Extract day name
df["Hari_Transaksi"] = df["Waktu_Transaksi"].dt.day_name()

# Weekend indicator
df["Is_Weekend"] = df["Hari_Transaksi"].isin(
    ["Saturday", "Sunday"]
)

print(df[
    [
        "Waktu_Transaksi",
        "Jam_Transaksi",
        "Hari_Transaksi",
        "Is_Weekend"
    ]
].head())
```

# =========================================
# 6. HANDLE OUTLIERS
# =========================================

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n")
print("=" * 50)
print("OUTLIER DETECTION (BOX PLOTS)")
print("=" * 50)

numerical_cols_for_outliers = [
    'Harga_Pesanan',
    'Waktu_Tunggu_Menit',
    'Jarak_Kirim_KM',
    'Rating_Pelanggan'
]

# Create subplots for each numerical column
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numerical_cols_for_outliers):
    sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Box Plot of {col}')
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
print("\n")
print("=" * 50)
print("OUTLIER REMOVAL (IQR METHOD)")
print("=" * 50)

print(f"Shape before outlier removal: {df.shape}")

# Define the columns where outliers will be removed
columns_to_clean = ['Harga_Pesanan', 'Waktu_Tunggu_Menit']

for column in columns_to_clean:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers for the current column
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

print(f"Shape after outlier removal: {df.shape}")
```

# =========================================
# 7. FINAL DATA VALIDATION
# =========================================

```python
print("\n")
print("=" * 50)
print("FINAL DATA VALIDATION")
print("=" * 50)

print(df.info())

print("\n")

print("=" * 50)
print("MISSING VALUES AFTER CLEANING")
print("=" * 50)

print(df.isnull().sum())
```

# =========================================
# 1. PEAK ORDER HOURS
# =========================================

```python
orders_per_hour = (
    df["Jam_Transaksi"]
    .value_counts()
    .sort_index()
)

plt.figure()

ax = sns.lineplot(
    x=orders_per_hour.index,
    y=orders_per_hour.values,
    marker="o",
    linewidth=3
)

plt.title(
    "Peak Order Hours",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Hour")
plt.ylabel("Total Orders")

for x, y in zip(
    orders_per_hour.index,
    orders_per_hour.values
):
    ax.text(
        x,
        y + 5,
        str(y),
        ha="center",
        fontsize=10
    )

sns.despine()

plt.show()
```

Order volume peaked during lunch hours (11:00–14:00), with the highest transaction volume recorded at 13:00 (839 orders). This pattern indicates a strong concentration of customer demand during midday periods. Improving operational readiness during peak hours could help maintain delivery performance and support revenue growth.

# =========================================
# 2. MENU CATEGORY DISTRIBUTION
# =========================================

```python
plt.figure()

category_counts = (
    df["Kategori_Menu"]
    .value_counts()
)

ax = sns.barplot(
    x=category_counts.values,
    y=category_counts.index
)

plt.title(
    "Most Ordered Menu Categories",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Total Orders")
plt.ylabel("Menu Category")

for i, value in enumerate(category_counts.values):
    ax.text(
        value + 10,
        i,
        str(value),
        va="center",
        fontsize=10
    )

sns.despine()

plt.show()
```

Ayam became the most frequently ordered menu category with 2,796 transactions, significantly outperforming other categories. Kopi also showed strong demand, indicating that customers frequently order beverages alongside main meals. Maintaining product availability and service quality for high-demand categories is essential to support customer satisfaction.

# =========================================
# 3. WAITING TIME DISTRIBUTION
# =========================================

```python
plt.figure()

sns.histplot(
    df["Waktu_Tunggu_Menit"],
    bins=30,
    kde=True
)

plt.axvline(
    df["Waktu_Tunggu_Menit"].mean(),
    linestyle="--",
    linewidth=2,
    label="Average Waiting Time"
)

plt.title(
    "Waiting Time Distribution",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Waiting Time (Minutes)")
plt.ylabel("Frequency")

plt.legend()

sns.despine()

plt.show()
```

Most orders were completed within 10–25 minutes, with an average waiting time of approximately 20 minutes. However, the distribution shows a long right tail, indicating that a portion of orders experienced significantly longer delivery times. These delayed orders may contribute to increased customer dissatisfaction and complaints.

# =========================================
# 4. WAITING TIME VS CUSTOMER RATING
# =========================================

```python
plt.figure()

sns.regplot(
    data=df,
    x="Waktu_Tunggu_Menit",
    y="Rating_Pelanggan",
    scatter_kws={"alpha": 0.4},
    line_kws={"linewidth": 3}
)

plt.title(
    "Waiting Time vs Customer Rating",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Waiting Time (Minutes)")
plt.ylabel("Customer Rating")

sns.despine()

plt.show()
```

Customer ratings show a gradual decline as waiting time increases, indicating a negative relationship between delivery duration and customer satisfaction. Although waiting time is not the sole factor affecting ratings, reducing delivery delays could help improve the overall customer experience.

# =========================================
# 5. COMPLAINT LEVEL ANALYSIS
# =========================================

```python
plt.figure()

sns.boxplot(
    data=df,
    x="Tingkat_Keluhan",
    y="Waktu_Tunggu_Menit"
)

plt.title(
    "Complaint Level vs Waiting Time",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Complaint Level")
plt.ylabel("Waiting Time (Minutes)")

sns.despine()

plt.show()
```

Orders associated with high complaint levels exhibited noticeably longer and more inconsistent waiting times compared to orders without complaints. This suggests that delivery delays and operational inconsistency may play a significant role in driving customer dissatisfaction.

# =========================================
# 6. PROMO EFFECTIVENESS ANALYSIS
# =========================================

```python
promo_rating = (
    df.groupby("Status_Promo")[
        "Rating_Pelanggan"
    ]
    .mean()
    .reset_index()
)

plt.figure()

ax = sns.barplot(
    data=promo_rating,
    x="Status_Promo",
    y="Rating_Pelanggan"
)

plt.title(
    "Promo vs Customer Rating",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Promo Status")
plt.ylabel("Average Rating")

for i, value in enumerate(
    promo_rating["Rating_Pelanggan"]
):
    ax.text(
        i,
        value + 0.03,
        f"{value:.2f}",
        ha="center",
        fontsize=11
    )

sns.despine()

plt.show()
```

Average customer ratings remained identical between promotional and non-promotional orders (4.17). This indicates that promotional campaigns alone may not significantly influence customer satisfaction, as customers continue to prioritize service quality and delivery experience over discounts.

# =========================================
# 7. MENU CATEGORY VS WAITING TIME
# =========================================

```python
category_waiting = (
    df.groupby("Kategori_Menu")[
        "Waktu_Tunggu_Menit"
    ]
    .mean()
    .sort_values(ascending=False)
)

plt.figure()

ax = sns.barplot(
    x=category_waiting.values,
    y=category_waiting.index
)

plt.title(
    "Average Waiting Time by Menu Category",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Average Waiting Time (Minutes)")
plt.ylabel("Menu Category")

for i, value in enumerate(
    category_waiting.values
):
    ax.text(
        value + 0.3,
        i,
        f"{value:.2f}",
        va="center",
        fontsize=10
    )

sns.despine()

plt.show()
```

Mie recorded the highest average waiting time among all menu categories, while Ayam showed the fastest average delivery performance. Although the differences were relatively small, longer preparation or delivery times for certain menu categories may contribute to higher customer complaints and operational inefficiencies.

# =========================================
# 8. CUSTOMER COMPLAINT DISTRIBUTION
# =========================================

```python
complaint_distribution = (
    df["Tingkat_Keluhan"]
    .value_counts()
)

plt.figure()

ax = sns.barplot(
    x=complaint_distribution.index,
    y=complaint_distribution.values
)

plt.title(
    "Customer Complaint Distribution",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Complaint Level")
plt.ylabel("Total Orders")

for i, value in enumerate(
    complaint_distribution.values
):
    ax.text(
        i,
        value + 20,
        str(value),
        ha="center",
        fontsize=10
    )

sns.despine()

plt.show()
```

# =========================================
# 9. WEEKEND VS WEEKDAY ANALYSIS
# =========================================

```python
weekend_analysis = (
    df.groupby("Is_Weekend")[
        "Waktu_Tunggu_Menit"
    ]
    .mean()
    .reset_index()
)

plt.figure()

ax = sns.barplot(
    data=weekend_analysis,
    x="Is_Weekend",
    y="Waktu_Tunggu_Menit"
)

plt.title(
    "Weekend vs Weekday Waiting Time",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Is Weekend")
plt.ylabel("Average Waiting Time")

for i, value in enumerate(
    weekend_analysis["Waktu_Tunggu_Menit"]
):
    ax.text(
        i,
        value + 0.3,
        f"{value:.2f}",
        ha="center",
        fontsize=10
    )

sns.despine()

plt.show()
```

Average waiting time during weekends was slightly higher compared to weekdays, suggesting increased operational pressure during peak demand periods. This indicates that additional operational capacity or resource optimization may be required to maintain service quality during weekends.
